### Principal Component Analysis (PCA) on Breast Cancer Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
## Load the dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python

print(cancer_dataset.DESCR)
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
scaled_data  # standarize data -3 - +3, mean = 0
```

```python
##  Applying PCA Algorithms
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # from given feature , make them into 2 features 
```

```python
data_pca=pca.fit_transform(scaled_data)
data_pca  # dimentionality Reduction 
```

```python
pca.explained_variance_
```

```python
import matplotlib.pyplot as plt

# Assuming data_pca and cancer_dataset are already defined

plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cancer_dataset['target'], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Cancer Dataset')
plt.colorbar(label='Target Classes')  # Optional: Add a colorbar for clarity
plt.show()
```
