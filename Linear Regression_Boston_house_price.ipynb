```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Import data 
df = pd.read_csv('C:/Users/user/Documents/GitHub/Machine-Learning-Projects/Boston_house_prices.csv')
```

```python
df
```

```python
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
df.columns=df.iloc[0]
df
```

```python
df=df.iloc[1:,:]
df=df.reset_index(drop=True)
df
```

```python
X_boston=df.iloc[:,:-1] # independent features 
Y_boston=df.iloc[:,-1] # dependent features 
X_boston
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_boston, Y_boston , train_size=0.80, test_size=0.20, random_state=42)
print('Train/Test Sets Sizes : ',X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()
```

```python
lr.fit(X_train,Y_train)
dt.fit(X_train,Y_train)
knn.fit(X_train,Y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = dt.predict(X_test)
y_pred3 = knn.predict(X_test)
```

```python

print("R^2 score for LR",r2_score(Y_test,y_pred1))
print("R^2 score for DT",r2_score(Y_test,y_pred2))
print("R^2 score for KNN",r2_score(Y_test,y_pred3))
```

```python
from sklearn.ensemble import BaggingRegressor

bag_regressor = BaggingRegressor(random_state=42)
bag_regressor.fit(X_train, Y_train)    
```

# Ridge Regression 

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Create an instance of Ridge
ridge = Ridge()

# Define the parameter grid
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 2, 5, 10, 20, 30]}

# Set up GridSearchCV
ridge_regression = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=5)

# Fit the model
ridge_regression.fit(X_train, Y_train)
```

```python
print(ridge_regression.best_estimator_)
print(ridge_regression.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Create an instance of Lasso
lasso = Lasso()

# Define the parameter lasso
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 2, 5, 10, 20, 30]}

# Set up GridSearchCV
lasso_regression = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=5)

# Fit the model
lasso_regression.fit(X_train, Y_train)
```

```python
print(lasso_regression.best_estimator_)
print(lasso_regression.best_score_)
```
